## 1.1) 샘플링된 플레이리스트별 곡 리스트 정리

```python
# 플레이리스트 아이디(id)와 수록곡(songs) 추출
plylst_song = sampled_train[['플레이리스트 ID', '곡 리스트']]

# unnest songs
# a. 플레이리스트id를 곡의 수(songs컬럼위 요소인 리스트 길이)만큼 반복 -> 플레이리스트id 컬럼(1차원 어레이)
# b. songs의 리스트의 곡id를 하나하나 갈라서 한 리스트에 담음 -> 곡id컬럼(1차원 어레이)
# a와 b 두 어레이를 세로로 붙인다. -> 3차원 어레이가 된다.
plylst_song_unnest = np.dstack(
    (
        np.repeat(plylst_song['플레이리스트 ID'].values, list(map(len, plylst_song['곡 리스트']))), 
        np.concatenate(plylst_song['곡 리스트'].values)
    )
)

# plylst_song_unnest어레이를 이용해 unnested 데이터프레임 생성 : plylst_song
plylst_song = pd.DataFrame(data = plylst_song_unnest[0], columns = plylst_song.columns)
# plylst_song_unnest[0] -> 가장 겉의 어레이를 벗겨서 data 매개변수에 전달
plylst_song['플레이리스트 ID'] = plylst_song['플레이리스트 ID'].astype(str) # int타입을 str타입으로 형변환
plylst_song['곡 리스트'] = plylst_song['곡 리스트'].astype(str) # int타입을 str타입으로 형변환

# unnest 객체 제거
del plylst_song_unnest # 어레이는 이제 필요없으니까 삭제
```

### 1.2.1) 중복되는 노래 제거 & 샘플링하고 각 노래별 태그 정리

```python
from tqdm import tqdm
from collections import OrderedDict

song_id_tag_list = []

for song_id in tqdm(sampled_songs):
    a = playlst_song_tag[playlst_song_tag['곡 리스트'] == song_id]
    a.reset_index(inplace=True)
    
    info = []

    for i in range(len(a)):
        b = a['태그 리스트'][i]
        for j in range(len(b)):
            info.append(b[j])
    
    res = list(OrderedDict.fromkeys(info))
    song_id_tag_list.append({'song_id': song_id, 'song_tag': res})
```

```python
indices = pd.DataFrame(song_genre_list_all.index, index = song_genre_list_all['곡 제목'])
```

```python
def get_recomendation(title, cossim = cosine_sim ):
    idx = indices.loc[title][0]
    scores = list(enumerate(cossim[idx])) # 코사인 유사도 점수 리스트

    # 점수를 기준으로 sort 하기
    scores = sorted(scores, key = lambda x:x[1], reverse=True)
    scores_top = scores[1:11] # Top 10 
    scores = [i[0] for i in scores_top] # index list
    
    a = pd.DataFrame(scores_top, columns=['id', 'scores'])
    z = song_genre_list_all.iloc[scores][['곡 제목', '아티스트 리스트', 'song_genre', 'song_tag']]
    result_of_rs = pd.merge(a, z, how = 'left', left_on = 'id', right_on = z.index)
    result_of_rs.drop('id', axis=1, inplace=True)
     
    return result_of_rs
```

```python
def text_based_recomendation(tfidf_model, tfidf_matrix, text):
    aaa = tfidf_model.transform([text]).toarray()
    t = cosine_similarity(tfidf_matrix, aaa)
    
    scores = list(enumerate(t)) # 코사인 유사도 점수 리스트

    # 점수를 기준으로 sort 하기
    scores = sorted(scores, key = lambda x:x[1], reverse=True)
    scores_top = scores[1:11] # Top 10 
    scores = [i[0] for i in scores_top] # index list
    
    a = pd.DataFrame(scores_top, columns=['id', 'scores'])
    z = song_genre_list_all.iloc[scores][['곡 제목', '아티스트 리스트', 'song_genre', 'song_tag']]
    result_of_rs = pd.merge(a, z, how = 'left', left_on = 'id', right_on = z.index)
    result_of_rs.drop('id', axis=1, inplace=True)
     
    return result_of_rs
```

# 3. 형태소 분석 기반 태그 유사도 비교-명사, 형용사, 동사

```python
# 예시
from konlpy.tag import Okt

voc = '형태소 란 문장을 구성하는 의미 요소 중 가장 작은 단위를 말한다.'
okt_pos = Okt().pos(voc, norm=True, stem=True)   # 형태소 분석 ( norm : 정규화 )


okt_filtering = [x for x, y in okt_pos if y in ['Noun', 'Adjective', 'Verb']]
print(okt_filtering)   
```

```python
from konlpy.tag import Okt

def okt_organize(text):
    okt_pos = Okt().pos(text, norm=True, stem=True)
    okt_filtering = [x for x, y in okt_pos if y in ['Noun', 'Adjective', 'Verb']]
    return okt_filtering
```

```python
song_genre_list_all['processed_song_tag'] = song_genre_list_all['song_tag'].apply(lambda x: okt_organize(x))
```

```python
def text_based_recomendation(tfidf_model, tfidf_matrix, text):
    aaa = tfidf_model.transform([text])#.toarray()
    feature_names = tfidf.get_feature_names()
    dtm_np = np.array(aaa.todense())

    aaa = pd.DataFrame(data = dtm_np, columns = feature_names)

    for stopword in stopwords:
        del aaa[stopword]
        
    t = cosine_similarity(tfidf_matrix, aaa)
    
    scores = list(enumerate(t)) # 코사인 유사도 점수 리스트

    # 점수를 기준으로 sort 하기
    scores = sorted(scores, key = lambda x:x[1], reverse=True)
    scores_top = scores[1:11] # Top 10 
    scores = [i[0] for i in scores_top] # index list
    
    a = pd.DataFrame(scores_top, columns=['id', 'scores'])
    z = song_genre_list_all.iloc[scores][['곡 제목', '아티스트 리스트', 'song_genre', 'song_tag', 'processed_song_tag']]
    result_of_rs = pd.merge(a, z, how = 'left', left_on = 'id', right_on = z.index)
    result_of_rs.drop('id', axis=1, inplace=True)
     
    return result_of_rs
```

```python
text = '겨울에 듣기 좋은 우울 노래' # Home 겨울 감성 카페 요코 비틀즈 존레논 팝 우울 타겟

new_text = okt_organize(text)
new_text = remove_dirty(str(new_text))

print(new_text)

text_based_recomendation(tfidf, sim_df, new_text)
```

```python
# train 노래 내에서 비교

def song_similarity(texts, df):
    all_df = []

    target = data_ft[data_ft['곡 제목'] == texts]['tag_vec_ft'].values[0][0].reshape(1, -1)
    for i in range(len(df)):
        sim = cosine_similarity(df['tag_vec_ft'][i], target)
        
        row = {'sim': sim[0][0],
             '곡 제목': df.loc[i, '곡 제목'],
             '아티스트': df.loc[i, '아티스트 리스트'],
             '장르': df.loc[i, 'song_genre'],
             '태그': df.loc[i, 'song_tag'],
             '전처리 태그': df.loc[i, 'processed_song_tag']}
        all_df.append(row)
        
    cos = pd.DataFrame(all_df)
    cos = cos.sort_values(by='sim', axis=0, ascending=False)
    
    return cos[1:11]
```

```python
# input text와 비교

def similarity(texts, df):
    all_df = []
    vec = ft.get_sentence_vector(texts).reshape((1,-1))
    
    for i in range(len(df)):
        #print(len(df))
        sim = cosine_similarity(vec, df['tag_vec_ft'][i])
        row = {'sim': sim[0][0],
             '곡 제목': df.loc[i, '곡 제목'],
             '아티스트': df.loc[i, '아티스트 리스트'],
             '장르': df.loc[i, 'song_genre'],
             '태그': df.loc[i, 'song_tag'],
             '전처리 태그': df.loc[i, 'processed_song_tag']}
        all_df.append(row)
        
    cos = pd.DataFrame(all_df)
    cos = cos.sort_values(by='sim', axis=0, ascending=False)
    
    return cos.head(10)
```

```python
text = '겨울에 듣는 감성깊은 우울 노래' # Home 겨울 감성 카페 요코 비틀즈 존레논 팝 우울 타겟
#text = '겨울 팝송 새벽'
#text = '잔잔한'

new_text = okt_organize(text)
new_text = remove_dirty(str(new_text))

print(new_text)

similarity(new_text, data_ft)
```
