```python
import re
 
def cleanText(readData):
 
    #텍스트에 포함되어 있는 특수 문자 제거
 
    text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', readData)
 
    return text
 

```

```python
gr = song['곡 장르 리스트']
sub = []
for n,i in enumerate(gr):
  a = cleanText(str(i))
  sub.append(a)

sub
  
```

```python
sub = pd.DataFrame(sub)
song['대분류str']=sub


song.head(10)
```

```python
genre_gn_all = pd.read_json('/content/drive/Shareddrives/BOAZ_Adv/melon_data/genre_gn_all.json', orient = 'index')
genre_gn_all = pd.DataFrame(genre_gn_all) 
genre_gn_all.head()
```

```python
genre_gn_all.reset_index(inplace = True)
genre_gn_all.columns = ['장르 코드', '장르']
genre_gn_all
```

```python
big_cat = genre_gn_all[genre_gn_all['장르 코드'].str[-2:] == '00']    #대분류만 뽑음
big_cat
```

```python
big_cat_gen = big_cat['장르 코드']     #대분류 코드만 뽑는다
big_cat_gen = big_cat_gen.reset_index(drop=True)

big_cat_gen
```

```python
발라드 = song[song['대분류str'].str.contains(big_cat_gen[0])]    #발라드 한번 출력해봄
발라드.sample(100)
```

```python
# print(big_cat[big_cat['장르 코드']=='GN3000'])
# print(big_cat_gen[29])
# 크리스마스 = song[song['대분류str'].str.contains(big_cat_gen[29])]    #크리스마스 장르 데이터 없음
# 크리스마스.sample(100)
```

```python
ans = []
for i in big_cat_gen:
  if i == 'GN3000':
    break
  print(i)
  sub = song[song['대분류str'].str.contains(i)]
  sub = list(sub.sample(100,random_state=924).index)
  for i in sub:
    ans.append(i)
  
ans  
```

```python
len(ans)
```

```python
sample_song = song.iloc[ans]
sample_song.to_csv('/content/drive/MyDrive/adv/sampling_song_data.csv',index =False)

```

```python
def get_tags(tag_list, threshold=0):
  if len(tag_list) > 0:
    return ' '.join([x for x in tag_list])
  else:
    return ''


train['tags_preprocessed'] = train['tags'].apply(lambda x: get_tags(x))

train
```

```python
from tqdm import tqdm
import time

result = ['']*len(song)
for i in tqdm(range(len(train)),desc="tqdm example"):
  for j in train['songs'][i]:
    if j in ans:
      result[j]+=train['tags_preprocessed'][i]
      # print(result[j])
```

```python
result_pd = pd.DataFrame(result,columns=['tags'])
```

```python
len(result_pd)
```

```python
result_pd = result_pd[result_pd['tags']!='']   #tag있는 경우만 뽑아서 저장
result_pd.reset_index(inplace=True)
result_pd
```
